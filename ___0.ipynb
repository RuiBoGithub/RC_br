# BE AWARE OF HYPER-PARAMETERS
`10.0` is inherited from the original EPFL PCNN example, where:

```python
D = temp / division_factor + previous_value
```

The division factor limits the neural-network update so that the predicted state does not change too sharply between timesteps. It is a numerical scaling choice, not a physical parameter.


````python

from pathlib import Path
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# ============================================================
# 1. ETHlib case
# ============================================================

RC_SIMULATOR = "ETHlib"

if RC_SIMULATOR not in sys.path:
    sys.path.insert(0, RC_SIMULATOR)

from _ import *
from _BR_ import *
from _zone_ import DebugZone as Zone
import supply_system
import emission_system


case = RCCase(
    year=2023,
    loc_json=Path("_json/location_params.json"),
    geo_json=Path("_json/geo_params.json"),
    default_json=Path("_json/default_params.json"),
    epw_path=Path("../src/_base/epw/ED-TMYx.2023.epw"),
    occupancy_profile_csv=Path("_data/default_occ.csv"),
    Zone=Zone,
    supply_system=supply_system,
    emission_system=emission_system,
)

case.occupancy_profile = (
    pd.read_csv(
        Path("_data/bldg_LoD.csv"),
        usecols=["Y"],
    )
    .iloc[:8760]
    .reset_index(drop=True)
)

case.occupancy_profile.columns = ["People"]

case.electricity_profile = (
    case.occupancy_profile.copy()
)


# ============================================================
# 2. Run ETHlib once
# ============================================================

result, eui, debug = run_model_case(
    case=case,
    sampled_params={},
    controller_mode="occupancy",
)

result = result.copy()

if not isinstance(result.index, pd.DatetimeIndex):
    timestamp_column = next(
        (
            column
            for column in [
                "Timestamp",
                "timestamp",
                "DateTime",
                "datetime",
            ]
            if column in result.columns
        ),
        None,
    )

    if timestamp_column is not None:
        result.index = pd.to_datetime(
            result[timestamp_column]
        )
    else:
        result.index = pd.date_range(
            start="2023-01-01 00:00:00",
            periods=len(result),
            freq="h",
        )


# ============================================================
# 3. Load measured heating and electricity
# ============================================================

METER_PATH = Path("_data/Metering_ISO.csv")


def to_series(values, name):
    if isinstance(values, pd.Series):
        series = values.copy()

    elif isinstance(values, pd.DataFrame):
        if values.shape[1] != 1:
            raise ValueError(
                f"{name}: expected one column, "
                f"found {values.shape[1]}."
            )

        series = values.iloc[:, 0].copy()

    else:
        raise TypeError(
            f"{name}: unsupported type "
            f"{type(values).__name__}."
        )

    series.name = name
    series.index = pd.to_datetime(series.index)

    return series.sort_index()


measured_heating = to_series(
    load_meter_heating(
        METER_PATH,
        "Main Heating",
        case.geometry["FLOOR_AREA"],
        "h",
    ),
    "Measured heating",
)

measured_electricity = to_series(
    load_meter_heating(
        METER_PATH,
        "Main Electricity",
        case.geometry["FLOOR_AREA"],
        "h",
    ),
    "Electricity use",
)


# ============================================================
# 4. Build the PCNN input DataFrame
# ============================================================

required_result_columns = [
    "HeatingEnergy",
    "OutsideTemp",
    "SolarGains",
]

missing_result_columns = [
    column
    for column in required_result_columns
    if column not in result.columns
]

if missing_result_columns:
    raise KeyError(
        "Missing ETHlib result columns: "
        f"{missing_result_columns}. "
        f"Available columns: {result.columns.tolist()}"
    )

RC_ENERGY_SCALE = 1.0

data = pd.concat(
    [
        measured_heating,
        measured_electricity,
        (
            pd.to_numeric(
                result["HeatingEnergy"],
                errors="coerce",
            )
            * RC_ENERGY_SCALE
        ).rename("RC heating"),
        pd.to_numeric(
            result["OutsideTemp"],
            errors="coerce",
        ).rename("Outside temperature"),
        pd.to_numeric(
            result["SolarGains"],
            errors="coerce",
        ).rename("Solar irradiation"),
    ],
    axis=1,
    join="inner",
).sort_index()

# Jan 20 to Mar 20 inclusive.
data = data.loc[
    "2023-01-20 00:00:00":
    "2023-03-20 23:00:00"
].copy()

# Calendar inputs, following the original example.
data["Month sin"] = np.sin(
    2 * np.pi * (data.index.month - 1) / 12
)

data["Month cos"] = np.cos(
    2 * np.pi * (data.index.month - 1) / 12
)

data["Weekday"] = data.index.dayofweek

data["Timestep sin"] = np.sin(
    2 * np.pi * data.index.hour / 24
)

data["Timestep cos"] = np.cos(
    2 * np.pi * data.index.hour / 24
)

data = data.dropna().copy()

if len(data) < 100:
    raise ValueError(
        f"Only {len(data)} complete hourly rows remain. "
        "Check timestamp alignment."
    )


# ============================================================
# 5. Adapt to the original EPFL one-step DataSet
# ============================================================

# The original DataSet uses:
# X = data.iloc[:-1]
# Y = target.iloc[1:]
#
# Therefore, known inputs for t+1 are shifted to row t.
shadow_data = pd.DataFrame(index=data.index)

shadow_data["Measured heating"] = (
    data["Measured heating"]
)

future_columns = [
    "RC heating",
    "Electricity use",
    "Outside temperature",
    "Solar irradiation",
    "Month sin",
    "Month cos",
    "Weekday",
    "Timestep sin",
    "Timestep cos",
]

for column in future_columns:
    shadow_data[column] = (
        data[column].shift(-1)
    )

# One sequence category only.
shadow_data["Case"] = 1.0

# Prevent a zero range during Case normalisation.
shadow_data.iloc[
    0,
    shadow_data.columns.get_loc("Case"),
] = 0.0

shadow_data = shadow_data.dropna().copy()

display(shadow_data.head())

print(
    "Data period:",
    shadow_data.index.min(),
    "to",
    shadow_data.index.max(),
)


# ============================================================
# 6. Import and patch the original EPFL workflow
# ============================================================

import EPFLpiml

# Original EPFL files may still import from "pcnn".
sys.modules.setdefault("pcnn", EPFLpiml)

import EPFLpiml.model as pcnn_model
from EPFLpiml.parameters import parameters
from EPFLpiml.module_shadow import PCNN as ShadowPCNN

pcnn_model.PCNN = ShadowPCNN
pcnn_model.check_initialization_physical_parameters = (
    lambda **kwargs: None
)

Model = pcnn_model.Model


# ============================================================
# 7. Data parameters
# ============================================================

data_params = dict(
    temperature_column=[
        "Measured heating",
    ],

    power_column=[
        "RC heating",
    ],

    case_column="Case",

    out_column="Outside temperature",

    neigh_column=None,

    inputs_D=[
        "Electricity use",
        "Solar irradiation",
        "Month sin",
        "Month cos",
        "Weekday",
        "Timestep sin",
        "Timestep cos",
    ],

    outside_walls=[0],

    neighboring_rooms=[],

    seed=0,
)


# ============================================================
# 8. Model parameters
# ============================================================

model_params = parameters(
    name="ETHlib_shadow_Jan20_Mar20",

    seed=0,

    overlapping_distance=4,

    warm_start_length=12,

    # Hourly data: maximum sequence = 3 days.
    maximum_sequence_length=24 * 3,

    minimum_sequence_length=24,

    learn_initial_hidden_states=True,

    decrease_learning_rate=False,

    learning_rate=0.0005,

    feed_input_through_nn=True,

    input_nn_hidden_sizes=[32],

    lstm_hidden_size=64,

    lstm_num_layers=2,

    layer_norm=True,

    batch_size=256,

    output_nn_hidden_sizes=[32],

    division_factor=10.0,

    device="cpu",

    verbose=2,
)

model_params.update(
    {
        # First 80% is development data:
        # 70% training + 10% validation.
        # Final 20% is testing.
        "validation_percentage": 0.10,
        "test_percentage": 0.20,

        # One sequence category.
        "heating": True,
        "cooling": False,

        # Editable shadow-PCNN settings.
        "include_previous_target_in_D": True,
        "include_physics_in_D": True,
        "include_out_column_in_D": True,

        "correction_mode": "relative_to_physics",

        "max_correction_fraction": 0.50,

        "minimum_correction_fraction": 0.05,

        "clamp_physics_nonnegative": True,

        "enforce_nonnegative_output": True,

        "initial_physics_weights": [1.0],

        "save": True,

        # Required by the original Model constructor.
        "initial_values_physical_parameters": {
            "a": [1.0],
            "b": [1.0],
            "c": [1.0],
            "d": [1.0],
        },
    }
)

# Compatibility with versions of parameters() that return [10.0].
model_params["division_factor"] = float(
    np.asarray(
        model_params["division_factor"],
        dtype=float,
    ).reshape(-1)[0]
)


# ============================================================
# 9. Build, train and show only the original-style outcome
# ============================================================

module = "PCNN"

# Set train=False after a model has been saved.
train = True

load = not train
load_last = True

piml = Model(
    data=shadow_data,
    module=module,
    model_params=model_params,
    data_params=data_params,
    load=load,
    load_last=load_last,
)

print(
    {
        "train_sequences": len(piml.train_sequences),
        "validation_sequences": len(
            piml.validation_sequences
        ),
        "test_sequences": len(piml.test_sequences),
    }
)

if train:
    piml.fit(150)


# ============================================================
# 10. Loss plot
# ============================================================

plt.plot(
    piml.train_losses,
    c="tab:blue",
    label="train",
)

plt.plot(
    piml.validation_losses,
    c="tab:green",
    ls="--",
    label="validation",
)

plt.plot(
    piml.test_losses,
    c="tab:red",
    ls="--",
    label="test",
)

if len(piml.validation_losses) > 0:
    plt.vlines(
        x=np.argmin(
            piml.validation_losses
        ),
        ymin=min(
            piml.train_losses
            + piml.validation_losses
            + piml.test_losses
        ),
        ymax=max(
            piml.train_losses
            + piml.validation_losses
            + piml.test_losses
        ),
        color="black",
        lw=2,
        label="best",
    )

plt.legend()
plt.grid()
plt.show()


# ============================================================
# 11. Reload the best saved model
# ============================================================

load = True
load_last = False

piml = Model(
    data=shadow_data,
    module=module,
    model_params=model_params,
    data_params=data_params,
    load=load,
    load_last=load_last,
)
````

````python
# Whole-period prediction
prediction_norm, measured_norm = piml.predict(
    data=(piml.X, piml.Y)
)

prediction_norm = (
    prediction_norm
    .detach()
    .cpu()
    .numpy()[0, :, 0]
)

measured_norm = (
    measured_norm
    .detach()
    .cpu()
    .numpy()[0, :, 0]
)

target_min = float(
    np.asarray(
        piml.model_kwargs["temperature_min"]
    ).reshape(-1)[0]
)

target_range = float(
    np.asarray(
        piml.model_kwargs["temperature_range"]
    ).reshape(-1)[0]
)

# Reverse the EPFL 0.1–0.9 normalisation
simulated = (
    (prediction_norm - 0.1)
    / 0.8
    * target_range
    + target_min
)

metered = (
    (measured_norm - 0.1)
    / 0.8
    * target_range
    + target_min
)

time = shadow_data.index[
    1:1 + len(simulated)
]

plt.figure(figsize=(12, 4))

plt.plot(
    time,
    metered,
    label="Metered",
)

plt.plot(
    time,
    simulated,
    label="PIML simulated",
)

plt.xlabel("Time")
plt.ylabel("Hourly heating energy use")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
````

````python
daily = (
    pd.DataFrame(
        {
            "Metered": metered,
            "Simulated": simulated,
        },
        index=time,
    )
    .dropna()
    .resample("D")
    .sum()
)

error = daily["Metered"] - daily["Simulated"]

daily_cvrmse = (
    np.sqrt(np.mean(error**2))
    / daily["Metered"].mean()
    * 100
)

daily_nmbe = (
    error.sum()
    / daily["Metered"].sum()
    * 100
)

print(f"Daily CVRMSE: {daily_cvrmse:.2f}%")
print(f"Daily NMBE: {daily_nmbe:.2f}%")
````
